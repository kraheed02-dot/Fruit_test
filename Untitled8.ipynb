```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Step 1: Custom Data
X = np.array([
    [150, 7],
    [120, 4],
    [170, 6],
    [130, 3],
    [140, 5],
    [160, 7]
])
y = np.array([0, 1, 0, 1, 2, 0])  # 0=Sweet, 1=Sour, 2=Neutral

# Step 2: Train KNN Model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)

# Step 3: Create a Grid of Points (Whole Area)
x_min, x_max = X[:, 0].min() - 10, X[:, 0].max() + 10  # Weight range
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1    # Color score range

xx, yy = np.meshgrid(np.arange(x_min, x_max, 1),
                     np.arange(y_min, y_max, 0.1))

# Step 4: Predict Each Point in Grid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Step 5: Plot the Decision Boundary
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.4, cmap=plt.cm.RdYlBu)

# Step 6: Plot Original Data Points
colors = ['red', 'green', 'blue']
labels = ['Sweet', 'Sour', 'Neutral']
for idx, class_label in enumerate(np.unique(y)):
    plt.scatter(X[y == class_label, 0], X[y == class_label, 1],
                c=colors[class_label], label=labels[class_label], edgecolor='k', s=500)

# Step 7: Labels
plt.xlabel("Weight (grams)")
plt.ylabel("Color Score")
plt.title("KNN Decision Boundary (Fruit Taste Prediction)")
plt.legend()
plt.grid(True)
plt.show()
```
